### E1. Sorting

Complete the code below to sort a list of random data in ascending order. Implement a compare-and-swap subroutine. 

```python
import random

ndata = 20
list_data = []
for i in range(ndata):
    list_data.append(random.random())
```

```python
def compare_and_swap(j, k):
    if list_data[j] > list_data[k]:
        list_data[j], list_data[k] = list_data[k], list_data[j]
    return
```

```python
print(list_data)
```

```python
for j in range(ndata-1):
    for k in range(j+1, ndata):
        compare_and_swap(j, k)
```

### E2. Factoring Decision Problem

Given integer $m$ and $l < m$, does $m$ have a non-trivial factor less than $l$? Implement an algorithm to solve the problem. 

```python
import random
max_num = 100   # you can set this to a larger number for more challenging cases
```

```python
def factoring_decision(m, l):
    assert l < m
    
    for i in range(2, l):
        if m % i == 0:
            return True
    
    return False
```

To complete the problem, you are supposed to write down a few extra lines to test your code. 

```python
m = random.randint(2, max_num)
l = random.randint(2, m - 1)
print(f'm = {m}, l = {l}')

print(f'factoring_decision({m}, {l}) = {factoring_decision(m, l)}')
```
